### **3. Описание динамики спроса**

Для начала посмотрим количество вакансий, опубликованных в каждый месяц. Напомним, что в датасете собраны вакансии за период 11.02.2024-07.05.2024.

```python
dataset.groupby(dataset['Дата публикации'].dt.month).size()
```

Создадим датасет для создания визуализации.

```python
monthly_data = dataset.groupby(dataset['Дата публикации'].dt.month).size().reset_index(name='Количество вакансий')
monthly_data
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_data, x='Дата публикации', y='Количество вакансий', marker='o')

plt.title('Динамика вакансий по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество вакансий')
plt.ylim(0, 4300)
plt.xticks(ticks=range(1, 7), labels=['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь'])

for i in range(len(monthly_data)):
    plt.text(monthly_data['Дата публикации'][i], monthly_data['Количество вакансий'][i] + 100,
             str(monthly_data['Количество вакансий'][i]), ha='center', va='bottom', fontsize=10)

plt.grid(True)

plt.show()
```

Как уже было сказано, бесплатная версия сервиса «РосНавык» дала возможность выгрузить вакансии лишь за 3 последних месяца. Так, нами были получены данные за период: 11.02.2024 – 07.05.2024. На графике выше представлена динамика вакансий по месяцам за данный период. Поскольку период охватил февраль и май лишь частично, мы ожидаемо получили, что в эти месяцы было опубликовано меньше всего вакансий, хотя некорректно это утверждать.

Поэтому, чтобы решить данную проблему, найдем среднее количество вакансий, опубликованных за день, для каждого месяца.

```python
dayly_data = dataset.groupby(dataset['Дата публикации'].dt.month).size().reset_index(name='Количество вакансий')

# количество дней в каждом месяце
dayly_data['Количество дней в месяце'] = dataset.groupby(dataset['Дата публикации'].dt.month)['Дата публикации'].nunique().values

# среднее кол-во вакансий за день в данном месяце
dayly_data['Среднее кол-во вакансий, опубликованных за день'] = round(dayly_data['Количество вакансий'] / dayly_data['Количество дней в месяце']).astype(int)

dayly_data
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=dayly_data, x='Дата публикации', y='Среднее кол-во вакансий, опубликованных за день', marker='o')

plt.title('Динамика среднего количества вакансий')
plt.xlabel('Месяц')
plt.ylabel('Количество вакансий')
plt.ylim(0, 200)
plt.xticks(ticks=range(1, 7), labels=['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь'])

for i in range(len(dayly_data)):
    plt.text(dayly_data['Дата публикации'][i], dayly_data['Среднее кол-во вакансий, опубликованных за день'][i] + 10,
             str(dayly_data['Среднее кол-во вакансий, опубликованных за день'][i]), ha='center', va='bottom', fontsize=10)

plt.grid(True)

plt.show()
```

**Выводы:**

В период майских праздников публиковалось меньше всего вакансий. В марте и апреле было размещено примерно равное количество вакансий, а выше всего спрос на аналитиков и дата-сайентистов наблюдался в феврале.

Кроме того, в целом спрос на специалистов в сфере аналитики и Data Science очень высок – каждый день размещается около сотни вакансий.

```python
pd.set_option('display.max_colwidth', None)
salary_not0[salary_not0['Заработная плата'] <= 15_000]
```

Итак, как мы поступим с зарплатами <= 15_000:

1. Удалим объявления со следующими индексами: 8531, 7459, 1959, 6446, 3853, 4017

  Либо по описанию так и не стало понятно, в какой валюте указана зарплата, либо вакансия не относится к аналитике (например, встретилось объявление на личного помощника).

2. Зарплата для объявлений с индексами 231, 1509, 3916, 7010, 7217 указана в рублях, ничего не делаем.

3. Остальные зарплаты указаны в долларах, поэтому по примерному курсу, который был весной, переводим остальные зарплаты в доллары.

```python
salary_not0.drop(index=[8531, 7459, 1959, 6446, 3853, 4017], inplace=True)

indexes_to_keep = [231, 1509, 3916, 7010, 7217]
salary_not0.loc[(salary_not0['Заработная плата'] <= 15_000) & (~salary_not0.index.isin(indexes_to_keep)), 'Заработная плата'] *= 90
```

Таким образом, вакансии с очень высокими заработными платами (выше 400­ 000 рублей) предполагаются для узкого круга высококвалифицированных специалистов.

Также мы рассчитали дополнительные метрики: среднее значение заработных плат равно 135­ 044 руб., а медианное — 108­ 000 рублей. Такая разница между средней и медианной зарплатами указывает на неравномерность распределения зарплат — существует много "выбросов", то есть вакансий с очень высокими заработными платами, которые завышают среднее значение. Поэтому для более точного представления об уровне заработной платы на рынке труда стоит смотреть на медианное значение, так как оно менее подвержено влиянию выбросов (аномальных значений).

Проверим гипотезу $H_0$ о нормальном распределении с помощью критерия Шапиро-Уилка:

```python
df_ws_1 = np.array(salary_not0[salary_not0['График работы']=='Вахтовый метод']['Заработная плата'])
df_ws_2 = np.array(salary_not0[salary_not0['График работы']=='Гибкий график']['Заработная плата'])
df_ws_3 = np.array(salary_not0[salary_not0['График работы']=='Неполный рабочий день']['Заработная плата'])
df_ws_4 = np.array(salary_not0[salary_not0['График работы']=='Полный день']['Заработная плата'])
df_ws_5 = np.array(salary_not0[salary_not0['График работы']=='Сменный график']['Заработная плата'])
df_ws_6 = np.array(salary_not0[salary_not0['График работы']=='Удаленная работа']['Заработная плата'])
```

```python
len(df_ws_1)
```

```python
len(df_ws_3)
```

```python
# print("Вахтовый метод:", stats.shapiro(df_ws_1[1]) # исключаем из анализа из-за недостаточности данных
print("Гибкий график p-value:", stats.shapiro(df_ws_2)[1])
# print("Неполный рабочий день p-value:", stats.shapiro(df_ws_3)[1]) # исключаем из анализа из-за недостаточности данных
print("Полный день p-value:", stats.shapiro(df_ws_4)[1])
print("Сменный график p-value:", stats.shapiro(df_ws_5)[1])
print("Удаленная работа p-value:", stats.shapiro(df_ws_6)[1])
```

Итак, исключим вакансии с вахтовым методом и неполным рабочим днем из анализа, поскольку таких объявлений недостаточно для проверки гипотезы, результаты будут ненадежны.

Критерий Шапиро-Уилка показал, что только для сменного графика работы мы не можем отклонить гипотезу о нормальном распределении заработных плат, для остальных графиков работы нулевая гипотеза отклоняется.

Теперь, чтобы проверить, есть ли статистически значимые различия в уровне заработных плат для разных графиков работы, применим непараметрический критерий сравнения средних нескольких независимых групп -- критерий Краскала-Уоллиса.

```python
stats.mstats.kruskalwallis(df_ws_2, df_ws_4, df_ws_5, df_ws_6)
```

**Вывод:**

Отклоняем нулевую гипотезу о равенстве средних значений ЗП для разных графиков работы. Значит различия в зарплатах у аналитиков и дата-сайентистов с разными графиками работы являются статистически значимыми. Таким образом, размер ЗП значительно зависит от графика работы.

---

Дополнительно применим непараметрический критерий Манна-Уитни, чтобы попарно сравнить зарплаты при разных графиках работы.

```python
work_schedules = [("Гибкий график", df_ws_2),
                  ("Полный день", df_ws_4),
                  ("Сменный график", df_ws_5),
                  ("Удаленная работа", df_ws_6)]

for i in range(len(work_schedules)):
    for j in range(i + 1, len(work_schedules)):
        schedule_1, data_1 = work_schedules[i]
        schedule_2, data_2 = work_schedules[j]
        stat, pvalue = stats.mstats.mannwhitneyu(data_1, data_2)
        print(f"{schedule_1} vs {schedule_2}: p-value = {pvalue:}")
```

**Вывод:**

Лишь для гибкого графика работы и полного рабочего дня нет статистически значимых различий в уровне ЗП.
